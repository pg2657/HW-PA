Modeling: KNN

```python
import numpy as np
from collections import Counter
from sklearn.metrics import accuracy_score, classification_report

class KNearestNeighbors:
    def __init__(self, n_neighbors=3):
        self.n_neighbors = n_neighbors

    def fit(self, X_train, y_train):
        self.X_train = np.array(X_train)
        self.y_train = np.array(y_train)

    def _euclidean_distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))

    def predict(self, X_test):
        predictions = [self._predict_single(x) for x in X_test]
        return np.array(predictions)

    def _predict_single(self, x):
        distances = [self._euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.n_neighbors]
        k_nearest_labels = [self.y_train[i] for i in k_indices]
        most_common = Counter(k_nearest_labels).most_common(1)
        return most_common[0][0]

knn = KNearestNeighbors(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("KNN Accuracy:", accuracy)
print("KNN Report:\n", classification_report(y_test, y_pred))
```

Hyperparameter Tuning: Rnadom Forest and Neural network

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder

pca = PCA(n_components=10)
X_train_PCA = pca.fit_transform(X_train)
X_test_PCA = pca.transform(X_test)
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
rf_params = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

rf_grid_search = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
rf_grid_search.fit(X_train_PCA, y_train_encoded)

print("Best parameters for Random Forest:", rf_grid_search.best_params_)
rf_best_model = rf_grid_search.best_estimator_
rf_predictions = rf_best_model.predict(X_test_PCA)

rf_accuracy = accuracy_score(y_test_encoded, rf_predictions)
rf_precision = precision_score(y_test_encoded, rf_predictions, average='binary')
rf_recall = recall_score(y_test_encoded, rf_predictions, average='binary')
rf_f1 = f1_score(y_test_encoded, rf_predictions, average='binary')
print(f"Random Forest - Accuracy: {rf_accuracy:.4f}, Precision: {rf_precision:.4f}, Recall: {rf_recall:.4f}, F1 Score: {rf_f1:.4f}")

nn_params = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'alpha': [0.0001, 0.001, 0.01]
}

nn_grid_search = GridSearchCV(MLPClassifier(max_iter=1000, random_state=42), nn_params, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
nn_grid_search.fit(X_train_PCA, y_train_encoded)

print("\nBest parameters for Neural Network:", nn_grid_search.best_params_)
nn_best_model = nn_grid_search.best_estimator_
nn_predictions = nn_best_model.predict(X_test_PCA)

nn_accuracy = accuracy_score(y_test_encoded, nn_predictions)
nn_precision = precision_score(y_test_encoded, nn_predictions, average='binary')
nn_recall = recall_score(y_test_encoded, nn_predictions, average='binary')
nn_f1 = f1_score(y_test_encoded, nn_predictions, average='binary')
print(f"Neural Network - Accuracy: {nn_accuracy:.4f}, Precision: {nn_precision:.4f}, Recall: {nn_recall:.4f}, F1 Score: {nn_f1:.4f}")
```
